```python
def orderDrink():

    menu = [{}, {"menu" :"아메리카노", "price": 2000}, {"menu" :"카페라떼", "price":2500}, {"menu" :"아이스티", "price":3000}]

    print("메뉴를 선택하세요")
    print("1. 아메리카노")
    print("2. 카페라떼")
    print("3. 아이스티")

    pick = int(input())
    order = menu[pick]["menu"]
    price = menu[pick]["price"]
    print()

    print("샷추가 여부를 선택해주세요 y/n")
    option1 = input()
    if option1 == "y" :
        order += ", 샷추가"
        price += 500
    print()

    print("ice/hot을 선택해주세요")
    print("1:ice, 2:hot")
    option2 = input()
    if option2 == "1" :
        order += f", ice"
    elif option2 == "2" :
        order += f", hot"
    print()

    print("size선택 : S/M/L")
    print("1:small, 2:midium, 3:large")
    option3 = input()
    if option3 == "1" :
        order += ", small"
    elif option3 == "2":
        order += ", midium"
        price += 500
    elif option3 == "3" :
        order += ", large"
        price += 1000
    print()
    return {"menu" : order, "price" : price}


def orderDesert():
    menu = [{}, {"menu" :"소금빵", "price": 3000}, {"menu" :"베이글", "price":4000}, {"menu" :"초콜릿 케잌", "price":5000}]

    print("메뉴를 선택하세요")
    print("1. 소금빵")
    print("2. 베이글")
    print("3. 초콜릿 케잌")

    pick = int(input())
    print()

    return menu[pick]


def main():
    # 메뉴,가격 형식의 딕셔너리
    orderMenu = []

    while True :
        print()
        print("카테고리를 선택해주세요")
        print("1 : 음료, 2 : 디저트")
        category = input()
        print()

        if category == "1" :
            orderMenu.append(orderDrink())
        elif category == "2" :
            orderMenu.append(orderDesert())
        else :
            print("해당 카테고리가 존재하지 않습니다. 다시 선택해주세요")
            continue

        print("1 : 결제하기, 2 : 장바구니 담기")
        action = input()
        if action == "2" : continue
        print()

        pay = ["신용카드", "카카오 페이", "현금"]
        print("결제수단 선택")
        print("0 : 신용카드, 1 : 카카오 페이, 2 : 현금")
        payType = int(input())
        print()


        print("#### 영수증 ####")
        totalPrice = 0

        for e in orderMenu :
            print(f"메뉴 : {e["menu"]} ,가격 : {e["price"]}")
            totalPrice += e["price"]

        print(f"결제금액 : {totalPrice}")
        print(f"결제수단 : {pay[payType]}")

        break


main()
```
