```python
import matplotlib.pyplot as plt  # 데이터 시각화를 위한 라이브러리 불러오기

# 데이터 정의
categories = ['A', 'B', 'C', 'D', 'E']  # 범주형 데이터 (예: 제품, 국가, 부서 등)
values = [10, 20, 15, 25, 30]  # 각 범주의 값 (예: 판매량, 점수, 개수 등)

# 막대 그래프 생성
plt.bar(categories, values, color='skyblue')  # 범주(categories)별로 values 값을 갖는 막대 생성, 색상 설정
plt.xlabel("Category")  # X축 라벨 설정
plt.ylabel("Value")  # Y축 라벨 설정
plt.title("Basic Bar Chart")  # 그래프 제목 설정

# 그래프 표시
plt.show()  # 그래프를 화면에 출력
```
