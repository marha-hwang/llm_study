- `"context_length"`ëŠ” ìœ„ì¹˜ ì„ë² ë”©(2ì¥ ì°¸ì¡°)ìœ¼ë¡œ ëª¨ë¸ì´ ë‹¤ë£° ìˆ˜ ìˆëŠ” ì…ë ¥ í† í°ì˜ ìµœëŒ€ ê°œìˆ˜ì…ë‹ˆë‹¤.

- `"qkv_bias"`ëŠ” ë©€í‹° í—¤ë“œ ì–´í…ì…˜ì˜ Linear ì¸µì—ì„œ ì¿¼ë¦¬, í‚¤, ê°’ì„ ê³„ì‚°í•  ë•Œ í¸í–¥ ìœ ë‹›ì„ ë„ì…í• ì§€ ì—¬ë¶€ë¥¼ ê²°ì •í•©ë‹ˆë‹¤. í˜„ëŒ€ì ì¸ LLMì˜ êµ¬ì„± ë°©ì‹ì„ ë”°ë¼ì„œ ì²˜ìŒì—ëŠ” ì´ ê°’ì„ ë¹„í™œì„±í™”í•˜ì§€ë§Œ ì˜¤í”ˆAIì˜ ì‚¬ì „ í›ˆë ¨ëœ GPT-2 ê°€ì¤‘ì¹˜ë¥¼ ëª¨ë¸ë¡œ ë¡œë“œí•  ë•Œ ì´ë¥¼ ë‹¤ì‹œ ì‚´í´ ë³´ê² ìŠµë‹ˆë‹¤(5ì¥ ì°¸ì¡°).

````python
import torch
import torch.nn as nn


class DummyGPTModel(nn.Module):
    def __init__(self, cfg):
        super().__init__()

        # í† í°ì„ë² ë”©, í¬ì§€ì…˜ ì„ë² ë”©, ë“œë¡­ì•„ì›ƒ ë¹„ìœ¨ ì´ˆê¸°í™”
        self.tok_emb = nn.Embedding(cfg["vocab_size"], cfg["emb_dim"])
        self.pos_emb = nn.Embedding(cfg["context_length"], cfg["emb_dim"])
        self.drop_emb = nn.Dropout(cfg["drop_rate"])

        # íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì„ n_layers ë§Œí¼ ì´ˆê¸°í™”
        self.trf_blocks = nn.Sequential(
            *[DummyTransformerBlock(cfg) for _ in range(cfg["n_layers"])])

        # Use a placeholder for LayerNorm
        self.final_norm = DummyLayerNorm(cfg["emb_dim"])
        self.out_head = nn.Linear(
            cfg["emb_dim"], cfg["vocab_size"], bias=False
        )

    def forward(self, in_idx):
        batch_size, seq_len = in_idx.shape
        tok_embeds = self.tok_emb(in_idx)
        pos_embeds = self.pos_emb(torch.arange(seq_len, device=in_idx.device))
        x = tok_embeds + pos_embeds
        x = self.drop_emb(x)
        x = self.trf_blocks(x)
        x = self.final_norm(x)
        logits = self.out_head(x)
        return logits


class DummyTransformerBlock(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        # ë”ë¯¸ í´ë˜ìŠ¤

    def forward(self, x):
        # ì´ ë¸”ë¡ì€ ì•„ë¬´ê²ƒë„ í•˜ì§€ ì•Šê³  ì…ë ¥ì„ ê·¸ëƒ¥ ë°˜í™˜í•©ë‹ˆë‹¤.
        return x


class DummyLayerNorm(nn.Module):
    def __init__(self, normalized_shape, eps=1e-5):
        super().__init__()
        # ì¸µ ì •ê·œí™” ì¸í„°í˜ì´ìŠ¤ë¥¼ í‰ë‚´ë‚´ê¸° ìœ„í•œ ë§¤ê°œë³€ìˆ˜

    def forward(self, x):
        # ì´ ë¸”ë¡ì€ ì•„ë¬´ê²ƒë„ í•˜ì§€ ì•Šê³  ì…ë ¥ì„ ê·¸ëƒ¥ ë°˜í™˜í•©ë‹ˆë‹¤.
        return x
````

## 4.2 ì¸µ ì •ê·œí™”ë¡œ í™œì„±í™” ì •ê·œí™”í•˜ê¸°

- LayerNorm([Ba et al. 2016](https://arxiv.org/abs/1607.06450))ì´ë¼ê³ ë„ ë¶ˆë¦¬ëŠ” ì¸µ ì •ê·œí™”ëŠ” ì‹ ê²½ë§ ì¸µì˜ í™œì„±í™”ë¥¼ í‰ê· ì´ 0ì´ê³  ë¶„ì‚°ì´ 1ì´ ë˜ë„ë¡ ì¡°ì •í•©ë‹ˆë‹¤.
- ì´ë¥¼ í†µí•´ í›ˆë ¨ì„ ì•ˆì •í™”í•˜ê³  ê°€ì¤‘ì¹˜ ìˆ˜ë ´ ì†ë„ë¥¼ ë†’ì¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ë‚˜ì¤‘ì— êµ¬í˜„í•˜ê² ì§€ë§Œ ì¸µ ì •ê·œí™”ëŠ” íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì˜ ë©€í‹° í—¤ë“œ ì–´í…ì…˜ ëª¨ë“ˆ ì „í›„ì— ì ìš©ë©ë‹ˆë‹¤. ë˜í•œ ìµœì¢… ì¶œë ¥ ì¸µ ì „ì—ë„ ì ìš©ë©ë‹ˆë‹¤.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch04_compressed/05.webp" width="600px">

- ì‘ì€ ì…ë ¥ ìƒ˜í”Œì„ ê°„ë‹¨í•œ ì‹ ê²½ë§ ì¸µì— í†µê³¼ì‹œì¼œ ì¸µ ì •ê·œí™”ì˜ ì‘ë™ ë°©ì‹ì„ ì•Œì•„ ë³´ì£ :

````python
torch.manual_seed(123)

# ë‹¤ì„¯ ê°œì˜ ì°¨ì›(íŠ¹ì„±)ì„ ê°€ì§„ ë‘ ê°œì˜ í›ˆë ¨ ìƒ˜í”Œì„ ë§Œë“­ë‹ˆë‹¤.
batch_example = torch.randn(2, 5)

# nn.Sequential : ë‘ê°œì˜ ì‹ ê²½ë§ ê³„ì¸µì„ ë¬¶ì–´ì„œ í•˜ë‚˜ì˜ ë¸”ë¡ìœ¼ë¡œ ë§Œë“¬
# nn.Linear(5, 6) ì…ë ¥ : 5, ì¶œë ¥ : 6 ì¸ ì‹ ê²½ë§ ê³„ì¸µì„ ë§Œë“¬, í–‰ë ¬ë¡œ í‘œí˜„í•˜ë©´ (5, 6)
# nn.ReLU() : í¬ê¸° ë²¡í„°ê°€ 6ì¸ ê° í–‰ë ¬ì— ReLUí•¨ìˆ˜ë¥¼ ì ìš©í•¨
layer = nn.Sequential(nn.Linear(5, 6), nn.ReLU())
out = layer(batch_example)
print(out)
````

- ë‘ ê°œì˜ ì…ë ¥ì— ëŒ€í•´ ê°ê° í‰ê· ê³¼ ë¶„ì‚°ì„ ê³„ì‚°í•´ ë³´ê² ìŠµë‹ˆë‹¤:

````python
# ë§ˆì§€ë§‰ ì°¨ì›ì˜ í‰ê· ì„ ê³„ì‚°í•˜ê¸° ìœ„í•œì½”ë“œ
# 2ì°¨ì› í–‰ë ¬ì—ì„œ í‰ê· ì„ ê³„ì‚°í•˜ë©´ 1ì°¨ì› í–‰ë ¬ë¡œ ê²°ê³¼ê°€ ì¶œë ¥ë˜ì§€ë§Œ, 
# keepdimì„ ì‚¬ìš©í•˜ì—¬ ê¸°ì¡´ ì°¨ì›ì„ ìœ ì§€ì‹œí‚¨ë‹¤.
mean = out.mean(dim=-1, keepdim=True)

# ë§ˆì§€ë§‰ ì°¨ì›ì˜ ë¶„ì‚°ì„ ê³„ì‚°í•˜ê¸° ìœ„í•œ ì½”ë“œ
var = out.var(dim=-1, keepdim=True)

print("í‰ê· :\n", mean)
print("ë¶„ì‚°:\n", var)
````

- ì •ê·œí™”ëŠ” ë‘ ì…ë ¥(í–‰)ì— ëŒ€í•´ ë…ë¦½ì ìœ¼ë¡œ ì ìš©ë©ë‹ˆë‹¤. dim=-1ì„ ì‚¬ìš©í•˜ë©´ í–‰ ì°¨ì›ì´ ì•„ë‹ˆë¼ ë§ˆì§€ë§‰ ì°¨ì›(ì´ ê²½ìš° íŠ¹ì„± ì°¨ì›)ì„ ë”°ë¼ ê³„ì‚°ì´ ìˆ˜í–‰ë©ë‹ˆë‹¤.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch04_compressed/06.webp" width="600px">

- í‰ê· ì„ ë¹¼ê³ , ë¶„ì‚°ì˜ ì œê³±ê·¼(í‘œì¤€í¸ì°¨)ìœ¼ë¡œ ë‚˜ëˆ„ë©´ ì…ë ¥ì„ ì—´(íŠ¹ì„±) ì°¨ì›ì„ ë”°ë¼ í‰ê· ì´ 0ì´ê³  ë¶„ì‚°ì´ 1ì´ ë˜ë„ë¡ ë§Œë“­ë‹ˆë‹¤:

````python
out_norm = (out - mean) / torch.sqrt(var)
print("ì •ê·œí™”ëœ ì¸µ ì¶œë ¥:\n", out_norm)

mean = out_norm.mean(dim=-1, keepdim=True)
var = out_norm.var(dim=-1, keepdim=True)
print("í‰ê· :\n", mean)
print("ë¶„ì‚°:\n", var)
````

- ê° ì…ë ¥ì˜ í‰ê· ì€ 0ì´ê³  ë¶„ì‚°ì€ 1ì…ë‹ˆë‹¤. ê²°ê³¼ë¥¼ ë³´ê¸° ì‰½ë„ë¡ íŒŒì´í† ì¹˜ì˜ ê³¼í•™ì  í‘œê¸°ë²•ì„ ëŒ ìˆ˜ ìˆìŠµë‹ˆë‹¤:

````python
torch.set_printoptions(sci_mode=False)
print("í‰ê· :\n", mean)
print("ë¶„ì‚°:\n", var)
````

- ìœ„ì—ì„œ ê° ì…ë ¥ì˜ íŠ¹ì„±ì„ ì •ê·œí™”í–ˆìŠµë‹ˆë‹¤.
- ì´ì œ ë™ì¼í•œ ì•„ì´ë””ì–´ë¥¼ ì‚¬ìš©í•´ `LayerNorm` í´ë˜ìŠ¤ë¥¼ êµ¬í˜„í•´ ë³´ì£ :

````python
class LayerNorm(nn.Module):
    def __init__(self, emb_dim):
        super().__init__()
        # ë‚˜ëˆ—ì…ˆì„ í•  ë•Œ 0ìœ¼ë¡œ ë‚˜ëˆ„ì–´ ë–¨ì–´ì§€ëŠ” ê²ƒì„ ë§‰ê¸° ìœ„í•´ ë§¤ìš° ì‘ì€ ê°’ ì •ì˜
        self.eps = 1e-5

        # ì´ˆê¸°ê°’ì´ 1ì¸ scaleíŒŒë¼ë¯¸í„°ë¥¼ ìƒì„±
        self.scale = nn.Parameter(torch.ones(emb_dim))

        # ì´ˆê¸°ê°’ì´ 0ì¸ shiftíŒŒë¼ë¯¸í„°ë¥¼ ìƒì„±
        self.shift = nn.Parameter(torch.zeros(emb_dim))

    def forward(self, x):
        mean = x.mean(dim=-1, keepdim=True)
        var = x.var(dim=-1, keepdim=True, unbiased=False) # correction=0

        # í‰ê· ì„ ë¹¼ê³ , ë¶„ì‚°ì˜ ì œê³±ê·¼(í‘œì¤€í¸ì°¨)ìœ¼ë¡œ í‰ê· ì´ 0ì´ê³  ë¶„ì‚°ì´ 1ì´ ë˜ë„ë¡ ë§Œë“¤ì–´ì§„ë‹¤.
        norm_x = (x - mean) / torch.sqrt(var + self.eps)

        # scaleê³¼ shiftëŠ” í•™ìŠµê°€ëŠ¥í•œ íŒŒë¼ë¯¸í„°
        return self.scale * norm_x + self.shift
````

**ìŠ¤ì¼€ì¼ ì¡°ì •ê³¼ ì´ë™**

- í‰ê· ì„ ë¹¼ê³ , ë¶„ì‚°ìœ¼ë¡œ ë‚˜ëˆ„ì–´ ì •ê·œí™”í•˜ëŠ” ê²ƒ ì´ì™¸ì— ë‘ ê°œì˜ í›ˆë ¨ ê°€ëŠ¥í•œ íŒŒë¼ë¯¸í„° `scale`ê³¼ `shift`ë¥¼ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.
- ì´ˆê¸° `scale`(1)ê³¼ `shift`(0)ì€ ì•„ë¬´ëŸ° ì˜í–¥ì„ ë¯¸ì¹˜ì§€ ëª»í•©ë‹ˆë‹¤. í•˜ì§€ë§Œ `scale`ê³¼ `shift`ê°€ í›ˆë ¨ ê°€ëŠ¥í•œ íŒŒë¼ë¯¸í„°ì´ê¸° ë•Œë¬¸ì— í›ˆë ¨ ê³¼ì •ì—ì„œ ë‘ íŒŒë¼ë¯¸í„°ë¥¼ ì¡°ì •í•˜ëŠ” ê²ƒì´ í›ˆë ¨ ì‘ì—…ì—ì„œ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í–¥ìƒì‹œí‚¨ë‹¤ê³  íŒë‹¨í•˜ëŠ” ê²½ìš° LLMì´ ìë™ìœ¼ë¡œ ì¡°ì •í•©ë‹ˆë‹¤.
- ì´ë¥¼ í†µí•´ ëª¨ë¸ì€ ì²˜ë¦¬í•˜ëŠ” ë°ì´í„°ì— ê°€ì˜ ì˜ ë§ëŠ” ìŠ¤ì¼€ì¼ ì¡°ì •ê³¼ ì´ë™ì„ í•™ìŠµí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ë¶„ì‚°ì˜ ì œê³±ê·¼ì„ ê³„ì‚°í•  ë•Œ ì‘ì€ ê°’(`eps`)ë¥¼ ë”í•©ë‹ˆë‹¤. ì´ëŠ” ë¶„ì‚°ì´ 0ì¼ ê²½ìš° 0 ë‚˜ëˆ—ì…ˆ ì˜¤ë¥˜ë¥¼ ë°©ì§€í•˜ê¸° ìœ„í•´ì„œì…ë‹ˆë‹¤.

**í¸í–¥ëœ ë¶„ì‚°**
- ìœ„ ë¶„ì‚° ê³„ì‚°ì—ì„œ `unbiased=False`ëŠ” $\frac{\sum_i (x_i - \bar{x})^2}{n}$ ì‹ìœ¼ë¡œ ë¶„ì‚°ì„ ê³„ì‚°í•œë‹¤ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤. `n`ì€ ìƒ˜í”Œ í¬ê¸°ì…ë‹ˆë‹¤(ì—¬ê¸°ì„œëŠ” íŠ¹ì„± ë˜ëŠ” ì—´ ê°œìˆ˜). ì´ ê³µì‹ì€ (ë¶„ëª¨ê°€ `n-1`ì¸) ë² ì…€ ë³´ì •(Bessel's correction)ì„ ì‚¬ìš©í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. ë”°ë¼ì„œ í¸í–¥ëœ ë¶„ì‚°ì„ ì¶”ì •í•©ë‹ˆë‹¤.
- LLMì—ì„œ ì„ë² ë”© ì°¨ì› `n`ì€ ë§¤ìš° í¬ë¯€ë¡œ `n`ê³¼ `n-1`ì„ ì‚¬ìš©í•˜ëŠ” ì°¨ì´ëŠ” ë¬´ì‹œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- í•˜ì§€ë§Œ GPT-2ê°€ ì •ê·œí™” ì¸µì— í¸í–¥ëœ ë¶„ì‚°ì„ ì‚¬ìš©í–ˆìœ¼ë¯€ë¡œ ë‚˜ì¤‘ì— ì‚¬ì „ í›ˆë ¨ëœ ê°€ì¤‘ì¹˜ë¥¼ ë¡œë“œí•  ë•Œ í˜¸í™˜ì„±ì„ ìœ„í•´ ë™ì¼í•œ ë°©ì‹ì„ ì ìš©í–ˆìŠµë‹ˆë‹¤.

## 4.3 GELU í™œì„±í™” í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ëŠ” í”¼ë“œ í¬ì›Œë“œ ë„¤íŠ¸ì›Œí¬ êµ¬í˜„í•˜ê¸°

- ì´ ì ˆì—ì„œ LLMì˜ íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì—ì„œ ì‚¬ìš©ë˜ëŠ” ì‘ì€ ì‹ ê²½ë§ ëª¨ë“ˆì„ êµ¬í˜„í•©ë‹ˆë‹¤.
- ë”¥ëŸ¬ë‹ì—ì„œëŠ” ReLU(Rectified Linear Unit) í™œì„±í™” í•¨ìˆ˜ê°€ ê°„ë‹¨í•˜ë©° ë‹¤ì–‘í•œ ì‹ ê²½ë§ êµ¬ì¡°ì—ì„œ íš¨ê³¼ì ì´ê¸° ë•Œë¬¸ì— ë„ë¦¬ ì‚¬ìš©ë©ë‹ˆë‹¤.
- LLMì—ì„œëŠ” ì „í†µì ì¸ ReLU ì™¸ì—ë„ ë‹¤ì–‘í•œ ì¢…ë¥˜ì˜ í™œì„±í™” í•¨ìˆ˜ê°€ ì‚¬ìš©ë©ë‹ˆë‹¤. ëŒ€í‘œì ì¸ ë‘ ê°œì˜ í•¨ìˆ˜ëŠ” GELU(Gaussian Error Linear Unit)ì™€ SwiGLU(Swish-Gated Linear Unit)ì…ë‹ˆë‹¤.
- GELUì™€ SwiGLUëŠ” ê°ê° ê°€ìš°ìŠ¤ ì˜¤ì°¨ í•¨ìˆ˜ì™€ ì‹œê·¸ëª¨ì´ë“œ GLU(gated linear unit)ì„ ì‚¬ìš©í•œ ë” ë³µì¡í•˜ê³  ë¶€ë“œëŸ¬ìš´ í™œì„±í™” í•¨ìˆ˜ì…ë‹ˆë‹¤. ê°„ë‹¨í•œ ReLUì™€ ë‹¬ë¦¬ ë”¥ëŸ¬ë‹ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í–¥ìƒì‹œí‚µë‹ˆë‹¤.

- GELU í™œì„±í™” í•¨ìˆ˜([Hendrycks and Gimpel 2016](https://arxiv.org/abs/1606.08415))ëŠ” ì—¬ëŸ¬ ë°©ë²•ìœ¼ë¡œ êµ¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì •í™•í•œ ì •ì˜ëŠ” GELU(x) = xâ‹…ğ›·(x)ì…ë‹ˆë‹¤. ì—¬ê¸°ì„œ ğ›·(x)ëŠ” í‘œì¤€ ê°€ìš°ìŠ¤ ëˆ„ì  ë¶„í¬ í•¨ìˆ˜(ê°€ìš°ìŠ¤ ì˜¤ì°¨ í•¨ìˆ˜)ì…ë‹ˆë‹¤.
- ì‹¤ì œë¡œëŠ” ê³„ì‚°í•˜ê¸° ì‰¬ìš´ ê·¼ì‚¬ì‹ìœ¼ë¡œ êµ¬í˜„í•©ë‹ˆë‹¤: $\text{GELU}(x) \approx 0.5 \cdot x \cdot \left(1 + \tanh\left[\sqrt{\frac{2}{\pi}} \cdot \left(x + 0.044715 \cdot x^3\right)\right]\right)
$ (ì›ë³¸ GPT-2 ëª¨ë¸ë„ ì»¤ë¸Œ í”¼íŒ…(curve fitting)ìœ¼ë¡œ ì°¾ì€ ì´ ê·¼ì‚¬ì‹ì„ ì‚¬ìš©í–ˆìŠµë‹ˆë‹¤).

````python
class GELU(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        return 0.5 * x * (1 + torch.tanh(
            torch.sqrt(torch.tensor(2.0 / torch.pi)) *
            (x + 0.044715 * torch.pow(x, 3))
        ))
````

````python
import matplotlib.pyplot as plt

gelu, relu = GELU(), nn.ReLU()

# ìƒ˜í”Œ ë°ì´í„°
x = torch.linspace(-3, 3, 100)
y_gelu, y_relu = gelu(x), relu(x)

plt.figure(figsize=(8, 3))
for i, (y, label) in enumerate(zip([y_gelu, y_relu], ["GELU", "ReLU"]), 1):
    plt.subplot(1, 2, i)
    plt.plot(x, y)
    plt.title(f"{label} activation function")
    plt.xlabel("x")
    plt.ylabel(f"{label}(x)")
    plt.grid(True)

plt.tight_layout()
plt.show()
````

- ì—¬ê¸°ì„œ ë³´ë“¯ì´ ReLUëŠ” ì–‘ìˆ˜ëŠ” ê·¸ëŒ€ë¡œ ì¶œë ¥í•˜ê³  ìŒìˆ˜ëŠ” ëª¨ë‘ 0ì„ ì¶œë ¥í•˜ëŠ” êµ¬ê°„ë³„ ì„ í˜• í•¨ìˆ˜(piecewise linear function)ì…ë‹ˆë‹¤.
- GELUëŠ” ë¶€ë“œëŸ¬ìš´ ë¹„ì„ í˜• í•¨ìˆ˜ë¡œ, ReLUì™€ ë¹„ìŠ·í•˜ì§€ë§Œ ëª¨ë“  ìŒìˆ˜ ê°’ì˜ ê·¸ë ˆì´ë””ì–¸íŠ¸ë¥¼ 0ìœ¼ë¡œ ë§Œë“¤ì§€ ì•ŠìŠµë‹ˆë‹¤(ëŒ€ëµ x = -0.75ì—ì„œëŠ” ê·¸ë ˆì´ë””ì–¸íŠ¸ê°€ 0ì´ ë©ë‹ˆë‹¤).
- ê·¸ë‹¤ìŒ LLMì˜ íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì— ì‚¬ìš©í•  ì‘ì€ ì‹ ê²½ë§ ëª¨ë“ˆì¸ `FeedForward`ë¥¼ êµ¬í˜„í•´ ë³´ì£ :

````python
print(GPT_CONFIG_124M["emb_dim"])

class FeedForward(nn.Module):
    def __init__(self, cfg):
        super().__init__()

        # í† í° ì„ë² ë”©ì˜ ì°¨ì›ì„ ì…ë ¥ìœ¼ë¡œ ë°›ì•„ì„œ 4ë°°ë¡œ ì¶œë ¥ì„ í™•ì¥ì‹œí‚¤ê³ 
        # GELUí•¨ìˆ˜ ì ìš©
        # 4ë°° í™•ì¥ëœ ì…ë ¥ì„ ë°›ì•„ì„œ ì›ë˜ ì°¨ì›ìœ¼ë¡œ ë‹¤ì‹œ ë³µê·€ì‹œí‚´
        self.layers = nn.Sequential(
            nn.Linear(cfg["emb_dim"], 4 * cfg["emb_dim"]),
            GELU(),
            nn.Linear(4 * cfg["emb_dim"], cfg["emb_dim"]),
        )

    def forward(self, x):
        return self.layers(x)
````

## 4.4 ìˆì»· ì—°ê²° ì¶”ê°€í•˜ê¸°

- ìŠ¤í‚µ ì—°ê²°(skip connection)ì´ë‚˜ ì”ì°¨ ì—°ê²°(residual connection)ì´ë¼ê³ ë„ ë¶€ë¥´ëŠ” ìˆì»· ì—°ê²°(shortcut connection) ì´ë©´ì˜ ê°œë…ì„ ì•Œì•„ ë³´ì£ .
- ì›ë˜ ìˆì»· ì—°ê²°ì€ ì»´í“¨í„° ë¹„ì „ ë¶„ì•¼ì˜ ì‹¬ì¸µ ì‹ ê²½ë§(êµ¬ì²´ì ìœ¼ë¡œ ì”ì°¨ ì‹ ê²½ë§(residual network))ì—ì„œ ê·¸ë ˆì´ë””ì–¸íŠ¸ ì†Œì‹¤ ë¬¸ì œë¥¼ ì™„í™”í•˜ê¸° ìœ„í•´ ì œì•ˆë˜ì—ˆìŠµë‹ˆë‹¤.
- ìˆì»· ì—°ê²°ì´ ê·¸ë ˆì´ë””ì–¸íŠ¸ê°€ í•œ ê°œ ì´ìƒì˜ ì¸µì„ ê±´ë„ˆ ë›°ì–´ ë„¤íŠ¸ì›Œí¬ì— íë¥¼ ìˆ˜ ìˆë„ë¡ ì§§ì€ ë‹¤ë¥¸ ê²½ë¡œë¥¼ ë§Œë“ ë‹¤ëŠ” ê²ƒì„ ë³´ì—¬ì¤ë‹ˆë‹¤.
- ì´ëŸ° ê²½ë¡œëŠ” í•œ ì¸µì˜ ì¶œë ¥ì„ ì´í›„ ì¸µì˜ ì¶œë ¥ì— ë”í•˜ì—¬ ë§Œë“¤ì–´ì§‘ë‹ˆë‹¤.
- ì‘ì€ ìƒ˜í”Œ ë„¤íŠ¸ì›Œí¬ë¡œ ì´ ì•„ì´ë””ì–´ë¥¼ ì„¤ëª…í•´ ë³´ì£ :

- ì½”ë“œë¡œ êµ¬í˜„í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

````python
class ExampleDeepNeuralNetwork(nn.Module):
    def __init__(self, layer_sizes, use_shortcut):
        super().__init__()
        self.use_shortcut = use_shortcut
        self.layers = nn.ModuleList([
            nn.Sequential(nn.Linear(layer_sizes[0], layer_sizes[1]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[1], layer_sizes[2]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[2], layer_sizes[3]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[3], layer_sizes[4]), GELU()),
            nn.Sequential(nn.Linear(layer_sizes[4], layer_sizes[5]), GELU())
        ])

    def forward(self, x):
        for layer in self.layers:
            # í˜„ì¬ ì¸µì˜ ì¶œë ¥ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
            layer_output = layer(x)
            # ìˆì»· ì—°ê²°ì„ ì ìš©í•  ìˆ˜ ìˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤.
            if self.use_shortcut and x.shape == layer_output.shape:
                x = x + layer_output
            else:
                x = layer_output
        return x


def print_gradients(model, x):
    # ì •ë°©í–¥ ê³„ì‚°
    output = model(x)
    target = torch.tensor([[0.]])

    # íƒ€ê¹ƒê³¼ ì¶œë ¥ì˜ ê°€ê¹Œìš´ ì •ë„ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ì†ì‹¤ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
    loss = nn.MSELoss()
    loss = loss(output, target)

    # ê·¸ë ˆì´ë””ì–¸íŠ¸ë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•œ ì—­ì „íŒŒ
    loss.backward()

    for name, param in model.named_parameters():
        if 'weight' in name:
            # ê°€ì¤‘ì¹˜ì˜ ê·¸ë ˆì´ë””ì–¸íŠ¸ì˜ í‰ê·  ì ˆëŒ“ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
            print(f"{name}ì˜ í‰ê·  ê·¸ë ˆì´ë””ì–¸íŠ¸ëŠ” {param.grad.abs().mean().item()}ì…ë‹ˆë‹¤.")
````

- ìˆì»· ì—°ê²°ì´ **ì—†ì„** ë•Œ ê·¸ë ˆë””ì–¸íŠ¸ ê°’ì„ ì¶œë ¥í•´ ë³´ì£ :

````python
layer_sizes = [3, 3, 3, 3, 3, 1]

sample_input = torch.tensor([[1., 0., -1.]])

torch.manual_seed(123)
model_without_shortcut = ExampleDeepNeuralNetwork(
    layer_sizes, use_shortcut=False
)
print_gradients(model_without_shortcut, sample_input)
````

- ê·¸ë‹¤ìŒ ìˆì»· ì—°ê²°ì´ **ìˆì„** ë•Œ ê·¸ë ˆì´ë””ì–¸íŠ¸ ê°’ì„ ì¶œë ¥í•©ë‹ˆë‹¤:

````python
torch.manual_seed(123)
model_with_shortcut = ExampleDeepNeuralNetwork(
    layer_sizes, use_shortcut=True
)
print_gradients(model_with_shortcut, sample_input)
````

## 4.5 ì–´í…ì…˜ê³¼ ì„ í˜• ì¸µì„ íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì— ì—°ê²°í•˜ê¸°

- ì´ ì ˆì—ì„œ ì´ì „ì— ë°°ìš´ ê°œë…ì„ ì†Œìœ„ íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì— ê²°í•©í•©ë‹ˆë‹¤.
- íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì€ ì´ì „ ì¸µì—ì„œ ë‹¤ë£¬ ì½”ì˜ ë©€í‹° í—¤ë“œ ì–´í…ì…˜ ëª¨ë“ˆê³¼ ì•ì„œ ë‹¤ë£¬ í”¼ë“œ í¬ì›Œë“œ ì‹ ê²½ë§ì„ ê²°í•©í•©ë‹ˆë‹¤.
- ë˜í•œ íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì€ ë“œë¡­ì•„ì›ƒê³¼ ìˆì»· ì—°ê²°ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.


````python
# ê¹ƒí—ˆë¸Œì—ì„œ previous_chapters.py íŒŒì¼ì„ ë‹¤ìš´ë¡œë“œí•©ë‹ˆë‹¤.
!wget https://bit.ly/43BW1j3 -O previous_chapters.py
````

````python
from gpt_module import MultiHeadAttention


class TransformerBlock(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.att = MultiHeadAttention(
            d_in=cfg["emb_dim"],
            d_out=cfg["emb_dim"],
            context_length=cfg["context_length"],
            num_heads=cfg["n_heads"],
            dropout=cfg["drop_rate"],
            qkv_bias=cfg["qkv_bias"])
        self.ff = FeedForward(cfg)
        self.norm1 = LayerNorm(cfg["emb_dim"])
        self.norm2 = LayerNorm(cfg["emb_dim"])
        self.drop_shortcut = nn.Dropout(cfg["drop_rate"])

    def forward(self, x):
        # ì–´í…ì…˜ ë¸”ë¡ì„ ìœ„í•œ ìˆì»· ì—°ê²°
        shortcut = x
        x = self.norm1(x)
        x = self.att(x)  # í¬ê¸°: [batch_size, num_tokens, emb_size]
        x = self.drop_shortcut(x)
        x = x + shortcut  # ì›ë˜ ì…ë ¥ì„ ë”í•©ë‹ˆë‹¤.

        # í”¼ë“œ í¬ì›Œë“œ ë¸”ë¡ì„ ìœ„í•œ ìˆì»· ì—°ê²°
        shortcut = x
        x = self.norm2(x)
        x = self.ff(x)
        x = self.drop_shortcut(x)
        x = x + shortcut  # ì›ë˜ ì…ë ¥ì„ ë”í•©ë‹ˆë‹¤.

        return x
````

## 4.6 GPT ëª¨ë¸ ë§Œë“¤ê¸°

- íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì„ ì´ ì¥ì˜ ì„œë‘ì—ì„œ ë§Œë“¤ì—ˆë˜ GPT êµ¬ì¡°ì— ì—°ê²°í•´ ë³´ì£ .
- íŠ¸ëœìŠ¤í¬ë¨¸ ë¸”ë¡ì€ ì—¬ëŸ¬ ë²ˆ ë°˜ë³µë©ë‹ˆë‹¤. 1ì–µ 2,400ë§Œ íŒŒë¼ë¯¸í„°ì˜ GPT-2 ëª¨ë¸ì˜ ê²½ìš° 12ë²ˆ ë°˜ë³µí•©ë‹ˆë‹¤:

- ì´ë¥¼ ì½”ë“œë¡œ êµ¬í˜„í•˜ë©´ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œ `cfg["n_layers"] = 12`ì…ë‹ˆë‹¤:

````python
class GPTModel(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.tok_emb = nn.Embedding(cfg["vocab_size"], cfg["emb_dim"])
        self.pos_emb = nn.Embedding(cfg["context_length"], cfg["emb_dim"])
        self.drop_emb = nn.Dropout(cfg["drop_rate"])

        self.trf_blocks = nn.Sequential(
            *[TransformerBlock(cfg) for _ in range(cfg["n_layers"])])

        self.final_norm = LayerNorm(cfg["emb_dim"])
        self.out_head = nn.Linear(
            cfg["emb_dim"], cfg["vocab_size"], bias=False
        )

    def forward(self, in_idx):
        batch_size, seq_len = in_idx.shape
        tok_embeds = self.tok_emb(in_idx)
        pos_embeds = self.pos_emb(torch.arange(seq_len, device=in_idx.device))
        x = tok_embeds + pos_embeds  # í¬ê¸° [batch_size, num_tokens, emb_size]
        x = self.drop_emb(x)
        x = self.trf_blocks(x)
        x = self.final_norm(x)
        logits = self.out_head(x)
        return logits
````

- 1ì–µ 2,400ë§Œ íŒŒë¼ë¯¸í„° ëª¨ë¸ì˜ ì„¤ì •ì„ ì‚¬ìš©í•´ ëœë¤í•œ ì´ˆê¸° ê°€ì¤‘ì¹˜ë¡œ GPT ëª¨ë¸ì„ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤:

````python
torch.manual_seed(123)
model = GPTModel(GPT_CONFIG_124M)

out = model(batch)
print("ì…ë ¥ ë°°ì¹˜:\n", batch)
print("\nì¶œë ¥ í¬ê¸°:", out.shape)
print(out)
````

## 4.7 í…ìŠ¤íŠ¸ ìƒì„±í•˜ê¸°

- ì•ì—ì„œ êµ¬í˜„í•œ GPTì™€ ê°™ì€ LLMì€ í•œ ë²ˆì— í•˜ë‚˜ì˜ ë‹¨ì–´ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch04_compressed/16.webp" width="500px">

- ë‹¤ìŒì— ë‚˜ì˜¤ëŠ” `generate_text_simple` í•¨ìˆ˜ëŠ” ê°„ë‹¨í•˜ê³  ë¹ ë¥´ê²Œ í…ìŠ¤íŠ¸ë¥¼ ìƒì„±í•˜ëŠ” ê·¸ë¦¬ë”” ë””ì½”ë”©(greedy decoding)ì„ êµ¬í˜„í•©ë‹ˆë‹¤.
- ê·¸ë¦¬ë”” ë””ì½”ë”©ì—ì„œëŠ” ê° ë‹¨ê³„ë§ˆë‹¤ ëª¨ë¸ì´ ê°€ì¥ ë†’ì€ í™•ë¥ ì„ ê°€ì§„ ë‹¨ì–´(ë˜ëŠ” í† í°)ì„ ë‹¤ìŒ ì¶œë ¥ìœ¼ë¡œ ì„ íƒí•©ë‹ˆë‹¤(ê°€ì¥ ë†’ì€ ë¡œì§“ì´ ê°€ì¥ ë†’ì€ í™•ë¥ ì— ëŒ€ì‘ë©ë‹ˆë‹¤. ë”°ë¼ì„œ ê¸°ìˆ ì ìœ¼ë¡œëŠ” ëª…ì‹œì ìœ¼ë¡œ ì†Œí”„íŠ¸ë§¤ë§¥ìŠ¤ í•¨ìˆ˜ë¥¼ ì ìš©í•  í•„ìš”ê°€ ì—†ìŠµë‹ˆë‹¤).
- ë‹¤ìŒ ì¥ì—ì„œ ì¡°ê¸ˆ ë” ê³ ê¸‰ ê¸°ë²•ì„ ì‚¬ìš©í•œ `generate_text` í•¨ìˆ˜ë¥¼ êµ¬í˜„í•˜ê² ìŠµë‹ˆë‹¤.
- ë‹¤ìŒ ê·¸ë¦¼ì€ GPT ëª¨ë¸ì´ ì£¼ì–´ì§„ ë¬¸ë§¥ì„ ê¸°ë°˜ìœ¼ë¡œ ë‹¤ìŒ í† í°ì„ ìƒì„±í•˜ëŠ” ë°©ë²•ì„ ë³´ì—¬ì¤ë‹ˆë‹¤.

````python
def generate_text_simple(model, idx, max_new_tokens, context_size):
    # idxëŠ” í˜„ì¬ ë¬¸ë§¥ì´ ë‹´ê¸´ (batch, n_tokens) í¬ê¸°ì˜ ì¸ë±ìŠ¤ ë°°ì—´ì…ë‹ˆë‹¤
    for _ in range(max_new_tokens):

        # í˜„ì¬ ë¬¸ë§¥ì´ ëª¨ë¸ì´ ì§€ì›í•˜ëŠ” ë¬¸ë§¥ í¬ê¸°ë¥¼ ì´ˆê³¼í•˜ë©´ ì˜ë¼ëƒ…ë‹ˆë‹¤.
        # ì˜ˆë¥¼ ë“¤ì–´, LLMì´ 5ê°œ í† í°ë§Œ ì§€ì›í•˜ê³  ë¬¸ë§¥ í¬ê¸°ê°€ 10ì´ë¼ë©´,
        # ë§ˆì§€ë§‰ 5ê°œ í† í°ë§Œ ë¬¸ë§¥ìœ¼ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.
        idx_cond = idx[:, -context_size:]

        # ì˜ˆì¸¡ì„ ë§Œë“­ë‹ˆë‹¤.
        with torch.no_grad():
            logits = model(idx_cond)

        # ë§ˆì§€ë§‰ íƒ€ì„ ìŠ¤í…ë§Œ ì‚¬ìš©í•˜ë¯€ë¡œ
        # (batch, n_token, vocab_size)ê°€ (batch, vocab_size)ê°€ ë©ë‹ˆë‹¤.
        logits = logits[:, -1, :]

        # í™•ë¥ ì„ ì–»ê¸° ìœ„í•´ ì†Œí”„íŠ¸ë§¥ìŠ¤ë¥¼ ì ìš©í•©ë‹ˆë‹¤.
        probas = torch.softmax(logits, dim=-1)  # (batch, vocab_size)

        # ê°€ì¥ ë†’ì€ í™•ë¥  ê°’ì„ ê°€ì§„ í•­ëª©ì˜ ì¸ë±ìŠ¤ë¥¼ ì–»ìŠµë‹ˆë‹¤.
        idx_next = torch.argmax(probas, dim=-1, keepdim=True)  # (batch, 1)

        # ì„ íƒí•œ ì¸ë±ìŠ¤ë¥¼ í˜„ì¬ ì‹œí€€ìŠ¤ì— ì¶”ê°€í•©ë‹ˆë‹¤.
        idx = torch.cat((idx, idx_next), dim=1)  # (batch, n_tokens+1)

    return idx
````

- `generate_text_simple`ëŠ” í•œ ë²ˆì— í•˜ë‚˜ì˜ í† í°ì„ ë§Œë“œëŠ” ë°˜ë³µì ì¸ ê³¼ì •ì„ êµ¬í˜„í•©ë‹ˆë‹¤.
